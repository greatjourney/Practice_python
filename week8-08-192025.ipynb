```python
class BankAccount:
    """make a bankaccount that has its balance and key number

    """
    def __init__(self, initial_balance,key_number):
        """
         >>> acc = BankAccount(10000, "secret")
         
        """
        self.__balance = initial_balance 
        self.__key_number = key_number

    def withdraw(self, amount,input_number):
        """If inputed number is correct,withdraw amount from the balance and return current balance

        >>> acc = BankAccount(10000, "secret")
        >>> acc.withdraw(4000, "secret")
        6000
        >>> acc.withdraw(5000, "some-other-password")
        'Incorrect password'
        >>> acc.withdraw(8000, "secret")
        'Insufficient funds'

        """
        if input_number != self.__key_number:
            return  "Incorrect password"
        if self.__balance >= amount:
            self.__balance -= amount
            return self.__balance
        else:
            return "Insufficient funds"

    def deposit(self, amount,input_number):
        """If inputed number is correct,deposit amount to the balance and return current balance

        >>> acc = BankAccount(10000, "secret")
        >>> acc.deposit(4000, "secret")
        14000
        >>> acc.withdraw(5000, "some-other-password")
        'Incorrect password'
        
        """

        if input_number != self.__key_number:
            return  "Incorrect password"
        self.__balance += amount
        return self.__balance
        
       
# help(BankAccount)
# help(BankAccount.withdraw)
# import doctest
# doctest.testmod(verbose=True)
```

## 課題2
クラスや関数を指定してdoctestする方法がわからず、今まで定義したクラス、関数全てのdoctestが出力されて非常に見づらかった。
doctestにprintの出力だけではなくreturnの返り値も書く必要があることがわからず、なぜfailedになるかわからなくて苦労した。

```python
import abc
class Observer:
    """abstract observer class"""
    @abc.abstractmethod
    def update(self):
        """update status"""

class Subject:
    """abstract subject class"""

    def __init__(self):
        self.observers =[]
    def attach(self, an_observer):
        """add observer to the list""" 
        self.observers.append(an_observer)

    def detach(self, an_observer):
        """remove observer from the list""" 
        self.observers.remove(an_observer)

    def notify(self):
        """call update() for all observers""" 
        for o in self.observers:
            o.update()

class ClockTimer(Subject):
    """a class manages time (self.hour, self.minute, and self.second)"""
    def __init__(self,hour,minute,second):
        super().__init__()
        self.hour = hour
        self.minute = minute
        self.second = second

    def tick(self):
        """
        >>> timer = ClockTimer(22, 54, 12)
        >>> timer.tick()
        13
        """
        self.second +=1
        if self.second == 60:
            self.second = 0
            self.minute +=1
            if self.minute == 60:
                self.minute = 0
                self.hour +=1
                if self.hour == 24:
                    self.hour = 0
        self.notify()
        return self.second

class ConsoleClock(Observer): 
    """display time in console
    
    >>> timer = ClockTimer(22, 54, 12)
    >>> cc1 = ConsoleClock(timer)
    >>> timer.tick()
    22:54:13
    13
    >>> cc2 = ConsoleClock(timer)
    >>> timer.tick()
    22:54:14
    22:54:14
    14
    >>> cc3 = ConsoleClock(timer)
    >>> timer.tick()
    22:54:15
    22:54:15
    22:54:15
    15
    >>> timer.detach(cc1)
    >>> timer.tick()
    22:54:16
    22:54:16
    16
    
    """ 
    def __init__(self, subject):
        """create ConsoleClock and attach to subject"""
        super().__init__() 
        self.subject = subject 
        self.subject.attach(self)

    def clock_to_string(self, clock):
        return "{:02d}:{:02d}:{:02d}".format(clock.hour, clock.minute, clock.second)

    def update(self): 
        print(self.clock_to_string(self.subject))

import doctest
doctest.testmod(verbose=True)
```
