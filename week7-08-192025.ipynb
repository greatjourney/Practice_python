## 課題1
変数hour,minute,secondを定め、secondの値を1増やすメソッドtick()では場合分けをし、
secondが60になったらminuteが1増え、minuteが60増えたらhourが1増え、houeが２４になったら全ての値が0になるようにした。

```python
class Counter:
    def __init__(self):
        self.__counter = 0
    
    def increment(self):
        self.__counter +=1

    def count(self):
        return self.__counter


class ClockTimer:
    def __init__(self,hour,minute,second):
        self.hour = hour
        self.minute = minute
        self.second = second

    def tick(self):
        self.second +=1
        if self.second == 60:
            self.second = 0
            self.minute +=1
            if self.minute == 60:
                self.minute = 0
                self.hour +=1
                if self.hour == 24:
                    self.hour = 0
        return self.second
```

## 課題2
値を記録する配列 self.__memoを定義し、長さは1000で全ての要素をFalseとした。あらかじめ0,1番目の要素には0,1を入れた。
fibでは、まずmemo[n]がFalseか数値かを判断し、数値であれば計算せずその値を返す。
そうでなければself.__memo[n] = self.fib(n-1)+self.fib(n-2)によって再帰でfib(n)の計算をする。その際も計算にはすでに記録されている値を用いることで実行回数をなるべく減らす。そうして計算が終わればmemo[n]の値を返す。こうすることで、fib(n)のfib()の実行回数はn以下になり、実行が大幅に効率化される。

```python
class Fib:
    def __init__(self):
        self.__counter = Counter()
        self.__memo = [False for i in range(1000) ]
        self.__memo[0] = 0
        self.__memo[1] = 1
        

    def fib(self, n):
        if self.__memo[n] != False:
            return self.__memo[n]
        self.__counter.increment() 
        if n == 0:
            return 0
        elif n ==1:
            return 1
        else:
            self.__memo[n] = self.fib(n-1)+self.fib(n-2)
            return self.__memo[n]

    def count(self):
        return self.__counter.count()

```

## 課題3
TitForTatPlayerにおいて、変数op_last_actionを定め、updateメソッドで相手の一つ前の行動を記録していき、playではop_last_actionがCooperateであれば自分もCooperate、Defectであれば自分もDefectする。こうすることで有力な戦力tit-for-tatをするPlayerが実装できる。
対戦結果を見ると、CooperatePlayerとの対戦ではどちらも協力し続け、DefectPlayerとの対戦では最初だけこちらが協力するが、あとはどちらも裏切り続けている。RandomPlayerとの対戦では相手の行動に応じて協力と裏切りを繰り返し、最終的な利益は、何度か試行したがいずれの場合も両者同じくらいになっている。AlternatePlayerとの対戦ではお互いが協力と裏切りを繰り返している。最終的な利益は両者同じである。

```python
Cooperate = 0
Defect = 1
def valid_action(act):
    return act == Cooperate or act == Defect 

Reward = [
    [ 2, 0 ],
    [ 3, 1 ],
]

class Player:
    def name(self): 
        pass

    def play(self):
        pass

    def update(self, my_action, op_action):
        pass
 
import random
class CooperatePlayer(Player):
    def name(self):
        return "CooperatePlayer"
    def play(self): 
        return Cooperate

class DefectPlayer(Player):
    def name(self):
        return "DefectPlayer"
    def play(self): 
        return Defect

class RandomPlayer(Player):
    def name(self):
        return "RandomPlayer"
    def play(self):
        return Cooperate if (random.randrange(2) == 0) else Defect

class AlternatePlayer(Player):
    def __init__(self):
        self.last_action = 0

    def name(self):
        return "AlternatePlayer"

    def update(self, my_action, op_action):
        self.last_action = my_action

    def play(self):
        return Cooperate if (self.last_action == 1) else Defect

class TitForTatPlayer (Player):
    def __init__(self):
        self.my_last_action = 0
        self.op_last_action = 0

    def name(self):
        return "TitForTatPlayer "
    
    def update(self, my_action, op_action):
        self.op_last_action = op_action

    def play(self):
        return Cooperate if (self.op_last_action == 0) else Defect


def play_one_game(player_a, player_b):
    act_a = player_a.play()
    act_b = player_b.play()
    if not valid_action(act_a): # 行動がルールに従っているかを確認
        raise ValueError
    if not valid_action(act_b): # 行動がルールに従っているかを確認
        raise ValueError
    reward_a = Reward[act_a][act_b]
    reward_b = Reward[act_b][act_a] 
    player_a.update(act_a, act_b) 
    player_b.update(act_b, act_a) 
    print(player_a.name(), " v.s. ", player_b.name()) # 結果を表示
    print(" actions: ", act_a, act_b)
    print(" rewards: ", reward_a, reward_b)

    return act_a, reward_a, act_b, reward_b

P = 1.0/8
def play_games(player_a, player_b):
    sum = [0, 0] # aとbの得点
    history = ["", ""] # a と b の選択を文字列にして保存
    while True:
        act_a, reward_a, act_b, reward_b = play_one_game(player_a, player_b) 
        sum[0] += reward_a
        sum[1] += reward_b
        history[0] += str(act_a)
        history[1] += str(act_b)
        if (random.random() < P): # 終了判定
            break
    print(sum, history)

pc = CooperatePlayer()
pd = DefectPlayer()
pr = RandomPlayer()
pa = AlternatePlayer()
pt = TitForTatPlayer()
play_games(pt,pc)
print()
pt = TitForTatPlayer()
play_games(pt,pd)
print()
pt = TitForTatPlayer()
play_games(pt,pr)
print()
pt = TitForTatPlayer()
play_games(pt,pa)

```
